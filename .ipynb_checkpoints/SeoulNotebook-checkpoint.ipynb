```python
# Loading Data

import pandas as pd

df = pd.read_csv('bikedata.csv')
```

```python
# Removing Unnecessary Columns

df.head()
df = df.drop('Unnamed: 0', axis = 1)
df = df.drop(['PDweek','DDweek'], axis = 1)
```

```python
df.columns
df.head()
```

```python
# Columns --> ['Duration', 'Distance', 'PLong', 'PLatd', 'DLong', 'DLatd', 'Haversine',
#              'Pmonth', 'Pday', 'Phour', 'Pmin', 'Dmonth', 'Dday', 'Dhour', 'Dmin',
#               'Temp', 'Precip', 'Wind', 'Humid', 'Solar', 'Snow', 'GroundTemp','Dust']
null_columns=df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```

```python
locdata = df.loc[: , ['PLong', 'PLatd', 'DLong', 'DLatd','Pmonth', 'Pday', 'Phour', 'Pmin', 'Dmonth', 'Dday', 'Dhour', 'Dmin']]
locdata.head()
```

```python
# Geographical representation of January Trips
# Need to find how many trips there are in each month
locdata['Pmonth'].value_counts().sort_values()
```
